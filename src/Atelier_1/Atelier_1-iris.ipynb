# Iris dataset

## Importations du dataset des iris

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import datasets

iris = datasets.load_iris()
```

```python
print(iris.DESCR)
```

```python
# iris_sepals = iris.data[:, 0]
# iris_petals = iris.data[:, 1]

# print(f"Iris sepals : \n{iris_sepals}\n")
# print(f"Iris petals : \n{iris_petals}\n")
```

## 2 - Create 2 variables to store data and labels

```python
X = iris.data
Y = iris.target
```

## 3 - Split the dataset

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=0.2)
```

## 4 - Import and train ML Algorithm

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rfc = RandomForestClassifier()
```

```python
rfc.fit(x_train, y_train)
```

## 5 - Make prediction

```python
ypred = rfc.predict(x_test)
```

## 6 - import and uses statistics metrics

```python
from sklearn.metrics import confusion_matrix, classification_report
```

```python
confusion_matrix(ypred, y_test)
```

```python
print(classification_report(ypred, y_test))
```

## 7 - Tricks hyperparameters

```python
rfc_2 = RandomForestClassifier(n_estimators=500)
```

```python
rfc_2.fit(x_train, y_train)
```

```python
ypred_2 = rfc_2.predict(x_test)
```

```python
confusion_matrix(ypred_2, y_test)
```

```python
print(classification_report(ypred_2, y_test))
```
